```python
# Load libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load datafile
filename = "datasets/datasets/DC_Properties.csv"
df = pd.read_csv(filename)
df.head(5)
```

```python
# List all column headings
df.columns
```

```python
# Find what type of number is PRICE
df['PRICE'].dtypes
```

```python
df['PRICE'].head(10)
```

```python
# Find max value of PRICE
df['PRICE'].max()
```

```python
# Bin data by price range
#Define bins as 0-249K, 250-499K, 500-749K, 750-999K, 1000-149000K
bins = [0, 250000, 500000, 750000, 1000000, 149000000]

# Create names for the four groups
group_names = ['0-249K', '250-499K', '500-749K', '750-999K', '1,000-149,000K']

#make new column with letter grades
df['Price Range'] = pd.cut(df['PRICE'], bins, labels=group_names)
df.head()
```

```python
df_histogram=pd.value_counts(df['Price Range']) # number of observations per letter grade
df_histogram
```

```python
plt.hist(df_histogram)
```
